```python
def my_func(x,y):
  if x>y:
    return x-y
  else:
    return x+y
arr1=[10,20,30,40,50]
arr2=[100,90,80,70,60]
vect_func=np.vectorize(my_func)
print("Array 1:",arr1)
print("Array 2:",arr2)
print("result:",vect_func(arr1,arr2))
```
